```python
import numpy as np
import pandas as pd
import datetime as dt
import pandas_datareader.data as web
import bs4 as bs
import urllib.request

start = dt.datetime(2019, 1, 15)
end = dt.datetime.now()
```

```python
df_old=pd.read_excel('Main_DF.xlsx')
df_old.set_index('Date',inplace=True)
df_old.tail(3)
```

```python
SRATE_USD_3M_Date = []
SRATE_USD_3M_Rate = []

source = urllib.request.urlopen('https://www.moex.com/ru/markets/currency/swaps.aspx?code=SRATE_USD_3M').read()
soup = bs.BeautifulSoup(source,'lxml')
table=soup.find('table',{'id':'ctl00_PageContent_ctrlGetSwaps_GridView1'})

for row in table.findAll('tr')[1:]:
    USD_3M_Date = row.findAll('td')[0].text
    SRATE_USD_3M_Date.append(USD_3M_Date)
    USD_3M_Rate= row.findAll('td')[1].text    
    SRATE_USD_3M_Rate.append(USD_3M_Rate)
    
USD_3M = pd.DataFrame({'Date':SRATE_USD_3M_Date,'Swap_3M':SRATE_USD_3M_Rate},columns=['Date', 'Swap_3M'] )
USD_3M['Date']=pd.to_datetime(USD_3M.Date,dayfirst=True)
USD_3M['Swap_3M']= USD_3M.Swap_3M.str.replace(",", ".")
USD_3M['Swap_3M']=USD_3M.Swap_3M.astype(float)
USD_3M.set_index('Date',inplace=True)
USD_3M=USD_3M.head(6)
USD_3M
```

```python
SRATE_USD_1Y_Date = []
SRATE_USD_1Y_Rate = []

source = urllib.request.urlopen('https://www.moex.com/ru/markets/currency/swaps.aspx?code=SRATE_USD_1Y').read()
soup = bs.BeautifulSoup(source,'lxml')
table=soup.find('table',{'id':'ctl00_PageContent_ctrlGetSwaps_GridView1'})

for row in table.findAll('tr')[1:]:
    USD_1Y_Date = row.findAll('td')[0].text
    SRATE_USD_1Y_Date.append(USD_1Y_Date)
    USD_1Y_Rate= row.findAll('td')[1].text    
    SRATE_USD_1Y_Rate.append(USD_1Y_Rate)
    
USD_1Y = pd.DataFrame({'Date':SRATE_USD_1Y_Date,'Swap_12M':SRATE_USD_1Y_Rate},columns=['Date', 'Swap_12M'] )
USD_1Y['Date']=pd.to_datetime(USD_1Y.Date,dayfirst=True)
USD_1Y['Swap_12M']= USD_1Y.Swap_12M.str.replace(",", ".")
USD_1Y['Swap_12M']=USD_1Y.Swap_12M.astype(float)
USD_1Y.set_index('Date',inplace=True)
USD_1Y=USD_1Y.head(6)
USD_1Y
```

```python
MOSPRIME_Date = []
MOSPRIME_ON_Rate = []
MOSPRIME_3M_Rate = []

source = urllib.request.urlopen('http://mosprime.com/archive').read()
soup = bs.BeautifulSoup(source,'lxml')
table=soup.find('table',{'class':'table table-bordered table-condensed arch-table rrr'})

for row in table.findAll('tr')[1:]:
    Date = row.findAll('th')[0].text
    MOSPRIME_Date.append(Date)
    
    MOSPRIME_ON= row.findAll('td')[0].text    
    MOSPRIME_ON_Rate.append(MOSPRIME_ON)
    
    MOSPRIME_3M= row.findAll('td')[5].text    
    MOSPRIME_3M_Rate.append(MOSPRIME_3M)
    
MOS_PRIME = pd.DataFrame({'Date':MOSPRIME_Date,'MOSPRIME_ON':MOSPRIME_ON_Rate,'MOSPRIME_3M':MOSPRIME_3M_Rate},columns=['Date', 'MOSPRIME_ON','MOSPRIME_3M'] )
MOS_PRIME['Date']=pd.to_datetime(MOS_PRIME.Date,dayfirst=True)
MOS_PRIME['MOSPRIME_ON']=MOS_PRIME.MOSPRIME_ON.astype(float)
MOS_PRIME['MOSPRIME_3M']=MOS_PRIME.MOSPRIME_3M.astype(float)
MOS_PRIME.set_index('Date',inplace=True)
MOS_PRIME=MOS_PRIME.head(6)
MOS_PRIME
```

```python
EUR_USD = web.get_data_moex('EURUSD000TOM', '2019-01-15', end)
EUR_USD=EUR_USD.loc[EUR_USD.BOARDID=='CETS']
EUR_USD.drop(['BOARDID', 'SECID', 'SHORTNAME', 'VOLRUR', 'OPEN', 'LOW', 'HIGH', 'NUMTRADES',  'WAPRICE'], 1, inplace=True)
EUR_USD.index.name= 'Date'
EUR_USD.sort_index(inplace=True,ascending=False) 
EUR_USD.rename(columns={'CLOSE':'EUR_USD'}, inplace=True)
EUR_USD=EUR_USD.head(6)
EUR_USD
```

```python
MOEXBC_New= web.get_data_moex('MOEXBC', '2019-05-22', end)
MOEXBC_New.drop(['BOARDID', 'SECID', 'SHORTNAME', 'NAME', 'OPEN', 'LOW', 'HIGH','DECIMALS', 'CURRENCYID',  'DIVISOR','VALUE','DURATION','CAPITALIZATION','YIELD'], 1, inplace=True)
MOEXBC_New.index.name= 'Date'
MOEXBC_New.index=pd.to_datetime(MOEXBC_New.index)
MOEXBC_New.rename(columns={'CLOSE':'MOEXBC'}, inplace=True)
MOEXBC_New.sort_index(inplace=True,ascending=False) 
MOEXBC_New=MOEXBC_New.head(6)
MOEXBC_New
```

```python
USD_RUB_New = web.get_data_moex('USD000UTSTOM', '2019-05-10', end)
USD_RUB_New=USD_RUB_New.loc[USD_RUB_New.BOARDID=='CETS']
USD_RUB_New.drop(['BOARDID', 'SECID', 'SHORTNAME', 'VOLRUR', 'OPEN', 'LOW', 'HIGH', 'NUMTRADES',  'WAPRICE'], 1, inplace=True)
USD_RUB_New.index.name= 'Date'
USD_RUB_New.rename(columns={'CLOSE':'USD_RUB'}, inplace=True)
USD_RUB_New.sort_index(inplace=True,ascending=False) 
USD_RUB_New=USD_RUB_New.head(6)
USD_RUB_New
```

```python
RGBITR_New= web.get_data_moex('RGBITR', '2019-05-22', end)
RGBITR_New.drop(['BOARDID', 'SECID', 'SHORTNAME', 'NAME', 'OPEN', 'LOW', 'HIGH','CLOSE','DECIMALS', 'CURRENCYID',  'DIVISOR','VALUE','DURATION','CAPITALIZATION'], 1, inplace=True)
RGBITR_New.index.name= 'Date'
RGBITR_New.index=pd.to_datetime(RGBITR_New.index)
RGBITR_New.rename(columns={'YIELD':'RGBITR_YIELD'}, inplace=True)
RGBITR_New.sort_index(inplace=True,ascending=False) 
RGBITR_New=RGBITR_New.head(6)
RGBITR_New
```

```python
USD_Repo = web.get_data_moex('MOEXREPOUSD', start, end)
USD_Repo.drop(['BOARDID', 'SECID', 'SHORTNAME', 'NAME', 'VALUE', 'OPEN', 'LOW', 'HIGH', 'DURATION', 'YIELD', 'DECIMALS', 'CAPITALIZATION', 'CURRENCYID', 'DIVISOR'], 1, inplace=True)
USD_Repo.rename(columns={'CLOSE':'USD_Repo'}, inplace=True)
USD_Repo.index.name= 'Date'
USD_Repo.index=pd.to_datetime(USD_Repo.index)
USD_Repo.sort_index(inplace=True,ascending=False) 
USD_Repo=USD_Repo.head(6)
USD_Repo
```

```python
my_data=[MOEXBC_New.MOEXBC, USD_RUB_New.USD_RUB,RGBITR_New.RGBITR_YIELD,USD_3M.Swap_3M, USD_1Y.Swap_12M, MOS_PRIME.MOSPRIME_ON, MOS_PRIME.MOSPRIME_3M, ]
my_data=pd.concat(my_data,axis=1)
# my_data.columns = ['Swap_12M','Swap_3M']
my_data.head(5)
```

```python
tickers = ['RSX','OIL','^IRX','CPER']
ETFs = pd.DataFrame()    

for ticker in tickers:
        ETF = web.get_data_yahoo(ticker, start, end)
        ETF.rename(columns={'Adj Close': ticker}, inplace=True)
        ETF.drop(['Open', 'High', 'Low', 'Close', 'Volume'], 1, inplace=True)
        if ETFs.empty:
            ETFs = ETF
        else:
            ETFs = ETFs.join(ETF, how='outer')
ETFs.rename(columns={'RSX': 'Russia_ETF', '^IRX':'Treas_4M', 'CPER':'Copper'}, inplace=True)
ETFs.tail(2)
```
